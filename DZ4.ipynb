## Задача 3:
- Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока.
- Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями.

```python
corr_matrix = df.corr()
corr_matrix = np.round(corr_matrix, 1)
```

```python
plt.figure(figsize=(15, 15))
sns.heatmap(corr_matrix, annot=True, linewidths=0.5, cmap='coolwarm');
```

Наибольшее влияние на цену имеют факторы:
- Жилая площать (0,7)
- Класс дома (0,7)
- Полезная площать (0,6)
- ???Жилая площать в 2015г (0,6)

```python
sns.jointplot(x=df['price'], y=df['sqft_living'], kind='reg')
plt.title('Price-Living sqr', loc= 'center');
```

Зависимость прямая, площадь равномерно влияет на стоимость, без серьезных отклонений.

```python
sns.jointplot(x=df['price'], y=df['grade'], kind='reg')
plt.title('Price-Grade');
```

Также прямая зависимость. Существуют выбросы при высоких оценках.

```python
sns.jointplot(x=df['price'], y=df['sqft_above'], kind='reg')
plt.title('Price-Sqr above');
```

Зависимость прямая. Существуют некоторые выбросы, скорее всего на них влияют и другие факторы помимо полезной площади.

```python
sns.jointplot(x=df['price'], y=df['sqft_living15'], kind='reg')
plt.title('Price-Living sqr', loc= 'center');
```

Зависимость есть, но по сравнению с обычным показателем жилой площади имеются выбросы и распределение более плотное.
